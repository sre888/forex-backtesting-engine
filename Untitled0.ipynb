```python
# Step 2: Download Forex data (EUR/USD)
pair = "EURUSD=X"   # FX symbol for EUR/USD in Yahoo Finance
start = "2015-01-01"
end = datetime.today().strftime("%Y-%m-%d")

df = yf.download(pair, start=start, end=end, progress=False)

# Display first few rows
print(f"Downloaded {len(df)} rows for {pair}")
df.head()
```

```python
# Plot the closing price
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Close'])
plt.title("EUR/USD Exchange Rate (Daily)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# Step 3A: Compute moving averages and trading signals

# Choose your moving average periods
short_window = 20   # fast-moving average (about 1 trading month)
long_window  = 50   # slow-moving average (about 2.5 months)

# Calculate the moving averages
df['SMA_short'] = df['Close'].rolling(short_window).mean()
df['SMA_long']  = df['Close'].rolling(long_window).mean()

# Generate trading signals
df['Signal'] = 0
df.loc[df['SMA_short'] > df['SMA_long'], 'Signal'] = 1   # Long when short > long

# Shift the signal to avoid lookahead bias (trade next day)
df['Position'] = df['Signal'].shift(1).fillna(0)

# Show last few rows
df.tail(10)
```

```python
# Step 3B: Visualize price with moving averages

plt.figure(figsize=(12,5))
plt.plot(df.index, df['Close'], label='Price', color='black')
plt.plot(df.index, df['SMA_short'], label=f'SMA {short_window}', color='blue')
plt.plot(df.index, df['SMA_long'], label=f'SMA {long_window}', color='orange')
plt.legend()
plt.title("EUR/USD with Short & Long Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")

# Shade the areas where we are in a long position
plt.fill_between(df.index, df['Close'].min(), df['Close'].max(),
                 where=(df['Position'] == 1), color='green', alpha=0.1,
                 label='Long Position')

plt.show()
```

```python
# Step 4A: Calculate daily returns and strategy performance

# Daily percentage change in price
df['Return'] = df['Close'].pct_change().fillna(0)

# Strategy return: position * daily return
df['Strategy_Return'] = df['Position'] * df['Return']

# Starting capital
initial_capital = 10_000  # $10,000 starting account
df['Equity'] = initial_capital * (1 + df['Strategy_Return']).cumprod()

# Show last few rows
df.tail(10)
```

```python
# Step 4B: Plot your equity curve

plt.figure(figsize=(12,5))
plt.plot(df.index, df['Equity'], color='green')
plt.title("Strategy Equity Curve (EUR/USD SMA Crossover)")
plt.xlabel("Date")
plt.ylabel("Account Value ($)")
plt.grid(True)
plt.show()
```

```python
# Step 4C: Compute simple performance metrics

final_value = df['Equity'].iloc[-1]
total_return = (final_value / initial_capital - 1) * 100
annualized_volatility = df['Strategy_Return'].std() * (252 ** 0.5)
sharpe_ratio = (df['Strategy_Return'].mean() / df['Strategy_Return'].std()) * (252 ** 0.5)

print(f"Starting Capital: ${initial_capital:,.2f}")
print(f"Final Capital:    ${final_value:,.2f}")
print(f"Total Return:     {total_return:.2f}%")
print(f"Annual Volatility:{annualized_volatility*100:.2f}%")
print(f"Sharpe Ratio:     {sharpe_ratio:.2f}")
```

```python
# Robust self-contained cell: flatten MultiIndex columns from yfinance and run the SMA backtest with costs

import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime

pair = "EURUSD=X"
start = "2015-01-01"
end = datetime.today().strftime("%Y-%m-%d")

print("Downloading", pair, "...")
df = yf.download(pair, start=start, end=end, progress=False)

# If yfinance returned MultiIndex columns, flatten them to single-level strings
if isinstance(df.columns, pd.MultiIndex):
    print("Flattening MultiIndex columns from yfinance...")
    # join levels with '_' but drop empty parts
    df.columns = ['_'.join([str(c) for c in col if str(c) != '']).strip('_') for col in df.columns]

# Now find a Close column robustly
close_candidates = [c for c in df.columns if c.lower() == 'close' or 'close' in c.lower()]
if not close_candidates:
    raise KeyError(f"No column containing 'Close' found in downloaded data. Columns: {df.columns.tolist()}")
# Prefer exact 'Close' if present
close_col = None
if 'Close' in df.columns:
    close_col = 'Close'
else:
    # prefer 'Adj Close' if present, else the first candidate
    if 'Adj Close' in df.columns:
        close_col = 'Adj Close'
    else:
        close_col = close_candidates[0]
        print("Using close column:", close_col)

# If we didn't rename when flattening, ensure we reference the right column name
if close_col not in df.columns:
    # in case candidate name contains spaces or underscores from flattening, try to find case-insensitive match
    matches = [c for c in df.columns if c.lower().endswith('close')]
    if matches:
        close_col = matches[0]
        print("Adjusted close_col to:", close_col)
    else:
        raise KeyError("Could not resolve a Close column after flattening. Columns: " + ", ".join(df.columns))

# Normalize to a single 'Close' column used by later code
if close_col != 'Close':
    df = df.rename(columns={close_col: 'Close'})

# Basic checks
if 'Close' not in df.columns:
    raise KeyError("After renaming, 'Close' still not present — aborting.")

# Sort and compute returns
df = df.sort_index()
df['Return'] = df['Close'].pct_change().fillna(0)

# Strategy: SMA crossover
short_window = 20
long_window = 50
df['SMA_short'] = df['Close'].rolling(short_window).mean()
df['SMA_long']  = df['Close'].rolling(long_window).mean()
df['Signal'] = 0
df.loc[df['SMA_short'] > df['SMA_long'], 'Signal'] = 1
df['Position'] = df['Signal'].shift(1).fillna(0)

# Capital and costs
initial_capital = 10_000.0
spread = 0.0002
commission_per_trade = 0.0

# Trade change and costs
df['Trade_Change'] = df['Position'].diff().abs().fillna(0)
df['Spread_Cost'] = np.where(df['Trade_Change'] > 0, spread / df['Close'], 0.0)
df['Commission_Cost'] = np.where(df['Trade_Change'] > 0, commission_per_trade / initial_capital, 0.0)

# Strategy returns and equity (net)
df['Strategy_Return'] = df['Position'] * df['Return']
df['Strategy_Return_Net'] = df['Strategy_Return'] - df['Spread_Cost'] - df['Commission_Cost']
df['Equity_Net'] = initial_capital * (1 + df['Strategy_Return_Net']).cumprod()

print("Done — flattened columns and computed strategy.")
print("Columns in df:", df.columns.tolist())
display(df[['Close','SMA_short','SMA_long','Position','Trade_Change','Spread_Cost','Strategy_Return_Net','Equity_Net']].tail(10))
```

```python
# ---------- Step 5B: Add Stop-Loss and Take-Profit ----------
stop_loss_pct = 0.01     # 1% stop loss
take_profit_pct = 0.02   # 2% take profit

# We'll track entry prices when we open a position
df['Entry_Price'] = np.nan
current_entry = None

for i in range(1, len(df)):
    # Detect new trade entry
    if df['Position'].iloc[i] == 1 and df['Position'].iloc[i-1] == 0:
        current_entry = df['Close'].iloc[i]
    elif df['Position'].iloc[i] == 0:
        current_entry = None

    # If position is open, check SL/TP
    if current_entry is not None:
        price_change = (df['Close'].iloc[i] - current_entry) / current_entry
        if price_change <= -stop_loss_pct or price_change >= take_profit_pct:
            # Exit trade
            df.loc[df.index[i:], 'Position'] = 0
            current_entry = None

df['Trade_Change'] = df['Position'].diff().abs().fillna(0)

# Recalculate returns and equity
df['Strategy_Return'] = df['Position'] * df['Return']
df['Strategy_Return_Net'] = df['Strategy_Return'] - df['Spread_Cost'] - df['Commission_Cost']
df['Equity_Net'] = initial_capital * (1 + df['Strategy_Return_Net']).cumprod()

print("✅ Stop-loss and take-profit rules added.")
display(df[['Close','Position','Trade_Change','Strategy_Return_Net','Equity_Net']].tail(10))
```

```python
# ---------- Step 5C: Add Position Sizing and Leverage ----------
risk_per_trade = 0.01    # risk 1% of current equity
leverage = 1.0            # 1x leverage

# Calculate dollar position size dynamically
df['Equity_Shift'] = df['Equity_Net'].shift(1).fillna(initial_capital)
df['Position_Size'] = df['Equity_Shift'] * risk_per_trade * leverage

# Adjust returns to reflect dynamic sizing
df['Position_Return'] = df['Position'] * df['Return'] * leverage
df['Strategy_Return_Net'] = df['Position_Return'] - df['Spread_Cost'] - df['Commission_Cost']

# New equity curve with dynamic sizing
df['Equity_Net'] = initial_capital * (1 + df['Strategy_Return_Net']).cumprod()

print("✅ Position sizing and leverage applied.")
display(df[['Close','Position','Position_Size','Strategy_Return_Net','Equity_Net']].tail(10))
```

```python
# ---------- Step 6: Final Visualization (After 5A + 5B + 5C) ----------

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Plot the strategy's equity curve
plt.plot(df.index, df['Equity_Net'], label='Strategy (Net with SL/TP & Costs)', linewidth=2)

# Plot Buy & Hold comparison
buy_hold_equity = 10_000 * (1 + df['Return']).cumprod()
plt.plot(df.index, buy_hold_equity, label='Buy & Hold', linestyle='--')

# Beautify chart
plt.title("EUR/USD SMA Strategy with SL/TP, Costs, and Position Sizing")
plt.xlabel("Date")
plt.ylabel("Equity ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------- Step 7: Performance Metrics ----------

import numpy as np

# Ensure no NaN values in returns
returns = df['Strategy_Return_Net'].dropna()

# 1️⃣ CAGR (Compound Annual Growth Rate)
total_years = (df.index[-1] - df.index[0]).days / 365.25
final_equity = df['Equity_Net'].iloc[-1]
CAGR = (final_equity / initial_capital) ** (1 / total_years) - 1

# 2️⃣ Sharpe Ratio (risk-adjusted return)
sharpe_ratio = np.sqrt(252) * returns.mean() / returns.std()

# 3️⃣ Max Drawdown
df['Cumulative_Max'] = df['Equity_Net'].cummax()
df['Drawdown'] = (df['Equity_Net'] / df['Cumulative_Max']) - 1
max_drawdown = df['Drawdown'].min()

# 4️⃣ Win Rate, Avg Gain/Loss
df['Trade_Returns'] = df['Strategy_Return_Net'] * 100
winning_trades = df[df['Trade_Returns'] > 0]
losing_trades = df[df['Trade_Returns'] < 0]
win_rate = len(winning_trades) / (len(winning_trades) + len(losing_trades)) * 100 if (len(winning_trades) + len(losing_trades)) > 0 else np.nan
avg_gain = winning_trades['Trade_Returns'].mean()
avg_loss = losing_trades['Trade_Returns'].mean()

# Display results
print("📈 Strategy Performance Metrics:")
print(f"Final Equity: ${final_equity:,.2f}")
print(f"CAGR: {CAGR*100:.2f}%")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Max Drawdown: {max_drawdown*100:.2f}%")
print(f"Win Rate: {win_rate:.2f}%")
print(f"Avg Gain: {avg_gain:.3f}%  |  Avg Loss: {avg_loss:.3f}%")
```

```python
# ---------- Step 8: SMA Optimization ----------
import warnings, itertools
warnings.filterwarnings("ignore")

results = []

# Define ranges to test
short_range = range(5, 30, 5)    # SMA short: 5,10,15,20,25
long_range  = range(30, 120, 10) # SMA long: 30,40,...110

for short_window, long_window in itertools.product(short_range, long_range):
    if short_window >= long_window:
        continue

    data = df.copy()

    # Recalculate signals
    data['SMA_short'] = data['Close'].rolling(short_window).mean()
    data['SMA_long']  = data['Close'].rolling(long_window).mean()
    data['Signal'] = 0
    data.loc[data['SMA_short'] > data['SMA_long'], 'Signal'] = 1
    data['Position'] = data['Signal'].shift(1).fillna(0)
    data['Strategy_Return'] = data['Position'] * data['Return']

    # Reapply cost model (same spread/commission)
    data['Trade_Change'] = data['Position'].diff().abs().fillna(0)
    data['Spread_Cost'] = np.where(data['Trade_Change'] > 0, spread / data['Close'], 0.0)
    data['Commission_Cost'] = np.where(data['Trade_Change'] > 0,
                                       commission_per_trade / initial_capital, 0.0)
    data['Strategy_Return_Net'] = data['Strategy_Return'] - data['Spread_Cost'] - data['Commission_Cost']

    # Compute performance
    total_years = (data.index[-1] - data.index[0]).days / 365.25
    final_equity = initial_capital * (1 + data['Strategy_Return_Net']).cumprod().iloc[-1]
    CAGR = (final_equity / initial_capital) ** (1 / total_years) - 1
    sharpe = np.sqrt(252) * data['Strategy_Return_Net'].mean() / data['Strategy_Return_Net'].std()
    max_dd = (data['Equity_Net'].cummax() - data['Equity_Net']).max() / data['Equity_Net'].cummax().max()

    results.append({
        'Short': short_window,
        'Long': long_window,
        'CAGR(%)': round(CAGR*100, 2),
        'Sharpe': round(sharpe, 2),
        'MaxDD(%)': round(-max_dd*100, 2)
    })

# Create results table
opt_df = pd.DataFrame(results).sort_values(by='CAGR(%)', ascending=False).reset_index(drop=True)

print("✅ Optimization Complete — Top 10 Results:")
display(opt_df.head(10))
```
